```python
print("Enter String")
_str=input()
print("U entered:")
print(_str)
print("length: ", len(_str))
_rev=_str [::-1]
print("reverse: ", _rev)
_nodup=''
for char in _str:
  if char not in _nodup:
    _nodup=_nodup+char
print("No duplicate: ", _nodup)
_vow='AaEeIiOoUu'
print("Vowel and Consonant")
_cntv=0
_cntc=0
for letter in _str:
  if letter in _vow:
    _cntv=_cntv+1
  else:
    _cntc=_cntc+1
print("Vowels: ", _cntv)
print("Consonant: ", _cntc)
```
